```python
df.info()
df.describe()
```

```python
#أفضل 10 موديلات من حيث عدد السيارات المدرجة.
top_models = df['model'].value_counts().head(10)
top_models
```

```python
#متوسط الاسعار لافضل 10 موديلات
avg_price_model = df.groupby('model')['price'].mean().sort_values(ascending=False).head(10)
avg_price_model
```

```python
#توزيع الاسعار مع تجاهل السيارات الفاخرة لوضوح الشكل
plt.figure(figsize=(10,5))
plt.hist(df['price'], bins=50, color='skyblue', edgecolor='black')
plt.title("Distribution of Car Prices")
plt.xlabel("Price")
plt.ylabel("Number of Cars")
plt.xlim(0, 50000) 
plt.show()
```

```python
#متوسط السعر بناء على سنة الصنع
avg_price_year = df.groupby('model_year')['price'].mean()

# رسم متوسط السعر بناء على السنة
plt.figure(figsize=(12,6))
plt.plot(avg_price_year.index, avg_price_year.values, marker='o', linestyle='-', color='green')
plt.title("Average Car Price by Year")
plt.xlabel("Model Year")
plt.ylabel("Average Price")
plt.grid(True)
plt.show()
```

```python
#نموذج العلاقة بين المسافة المقطوعه والسعر
plt.figure(figsize=(10,6))
plt.scatter(df['odometer'], df['price'], alpha=0.3, color='orange')
plt.title("Price vs Odometer")
plt.xlabel("Odometer (miles)")
plt.ylabel("Price")
plt.xlim(0, 500000) # لتجنب السيارات الفاخرة جداً
plt.ylim(0, 50000) #  العلاقة بشكل أوضح
plt.grid(True)
plt.show()
```

```python
#متوسط السعر بناء على حالة السيارة
avg_price_condition = df.groupby('condition')['price'].mean().sort_values(ascending=False)
avg_price_condition

# رسم توضيحي شريطي
plt.figure(figsize=(8,5))
avg_price_condition.plot(kind='bar', color='purple', edgecolor='black')
plt.title("Average Price by Car Condition")
plt.xlabel("Condition")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.show()
```

```python
# عدد السيارات بناء على نوع الوقود
fuel_counts = df['fuel'].value_counts()

# رسم توضيحي شريطي
plt.figure(figsize=(8,5))
fuel_counts.plot(kind='bar', color='teal', edgecolor='black')
plt.title("Number of Cars by Fuel Type")
plt.xlabel("Fuel Type")
plt.ylabel("Number of Cars")
plt.xticks(rotation=0)
plt.show()
```

```python

#تحويل العمود الدفع الرباعي الى قيمة حقيقة
df['is_4wd'] = df['is_4wd'].replace({True: 'Yes', False: 'No', 'True':'Yes', 'False':'No'})

# متوسط السعر بناء على الدفع الرباعي
avg_price_4wd = df.groupby('is_4wd')['price'].mean()

# رسم توضيحي شريطي
plt.figure(figsize=(6,5))
avg_price_4wd.plot(kind='bar', color='orange', edgecolor='black')
plt.title("Average Price: 4WD vs Non-4WD")
plt.xlabel("4WD")
plt.ylabel("Average Price")
plt.xticks(rotation=0)
plt.show()
```

```python
#ملخص التحليل:
#-السيارات ذو الدفع الرباعي أغلى السيارات
#-السيارات في حالة ممتازة(excellent) الاعلى سعرا
#-السيارات التي قطعت مسافات اقل هي الاعلى سعرا
#البنزين هو الاكثر شائعا واستخداما
#موديل السيارات التالية (sedan,SUV) هي الاكثر انتشارا
```
